Now we'll link those codes to another table containing the team conference, the team stadium, and other information

```python
link = 'https://en.wikipedia.org/wiki/National_Football_League#Teams'
```

```python
table = pd.read_html(link, header=0)[3]
```

```python
table = table.rename(columns={'Division[63]':'Conference Division',
                      'Club[63]':'Team',
                      'Stadium[64]':'Stadium'
                     })
```

```python
replace_dict = dict({
                    'Washington Football Team*[G]':'Washington Football Team',
                    'Chicago Bears*†':'Chicago Bears',
                    'Detroit Lions*':'Detroit Lions',
                    'Arizona Cardinals*†':'Arizona Cardinals',
                    'Los Angeles Rams*':'Los Angeles Rams',
                    'Indianapolis Colts*':'Indianapolis Colts',
                    'Tennessee Titans*':'Tennessee Titans',
                    'Kansas City Chiefs*':'Kansas City Chiefs',
                    'Las Vegas Raiders*':'Las Vegas Raiders',
                    'Los Angeles Chargers*':'Los Angeles Chargers'
                    })
```

```python
afc_table = table[1:17]
nfc_table = table[18:34]
```

```python
afc_table['Conference Division'] = 'AFC ' + afc_table['Conference Division'].astype(str)
nfc_table['Conference Division'] = 'NFC ' + nfc_table['Conference Division'].astype(str)
dfs = [afc_table, nfc_table]
table = pd.concat(dfs)
```

```python
table
```

```python
table = table.rename(columns={'Division[63]':'Conference Division',
                      'Club[63]':'Team',
                      'Stadium[64]':'Stadium'
                     })
```

```python
table = table.replace(replace_dict)
```

```python
table = table.merge(teamcode_df, on='Team', how='outer').drop(columns={'Capacity', 'First season[65]', 'Head coach', 'Coordinates'})
```
